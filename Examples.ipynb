```python
import numpy as np
import pandas as pd
from physt import h1, h2, h3

np.random.seed(42)

%matplotlib inline
```

```python
# Generate data to use
NUM = 100000
ekin = np.random.normal(60, 5, NUM)
x = np.random.normal(0, 3, NUM) + (ekin - 60) ** 2 / 20
y = np.random.normal(0, 10, NUM) - x / 3

df = pd.DataFrame({
    "ekin": ekin,
    "x": x,
    "y": y
})
df.to_csv("protons.csv")
```

```python
particles = pd.read_csv("protons.csv")
h = h1(particles["ekin"], title="Energy distribution of protons")
h
```

```python
hx = h2(particles["x"], particles["y"], "fixed_width", 5, adaptive=True)
hx.plot(figsize=(5, 5), show_zero=False, show_colorbar=False, cmap="rainbow", write_to="pre.png", title="Pre")
hx << (43.4, -27.5)
hx.plot(figsize=(5, 5), show_zero=False, show_colorbar=False, cmap="rainbow", write_to="post.png", title="Post")
```
